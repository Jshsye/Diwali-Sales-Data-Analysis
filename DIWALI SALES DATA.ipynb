```python
import chardet

# Detect encoding
with open("C:/Users/pavan/OneDrive/Desktop/PYTHON PROJECT/Diwali Sales Data.csv", 'rb') as file:
    result = chardet.detect(file.read())
print(result['encoding'])  # Check detected encoding

# Use detected encoding to read CSV
df = pd.read_csv("C:/Users/pavan/OneDrive/Desktop/PYTHON PROJECT/Diwali Sales Data.csv", encoding=result['encoding'])

```

```python
df=df.drop(['Status' , 'unnamed1'], axis=1)
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
import seaborn as se
```

```python
 ax=se.countplot(x='Gender', data=df)   #for Graph Represent
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
se.barplot(x='Gender', y= 'Amount' , data=sales)
```

```python
# From the above Graph we can know that female purchases
# are more than as compare to male and female peoples are spending more more money as compare to man
```

```python
se.countplot(data=df, x = 'Age Group' , hue = 'Gender')
```

```python
age=se.countplot(data=df, x = 'Age Group' , hue = 'Gender')
for bars in age.containers:
    age.bar_label(bars)
```

```python
# From the Above Graph We Got the Most People Purchased is At 
# the age of 26-35 and in that Female Peoples Purchases more as compare to male people
```

```python
sales=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
se.barplot(x='Age Group', y= 'Amount' , data=sales)
```

```python
# From the above chart as we know that most of the buyers are from 26-35 age group and especially female 
```

```python
sale_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
se.set(rc={'figure.figsize':(25,5)})
se.barplot(data=sale_state, x='State', y='Orders')
```

```python
sale_state=df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
se.set(rc={'figure.figsize':(25,5)})
se.barplot(data=sale_state, x= 'State', y = 'Amount')
```

```python
# From the above data we got that the highest sales are made in 
#  uttar pradesh, maharastra, karnataka so the more sale must be done with this states
```

```python
 ax=se.countplot(x='Marital_Status', data=df)   #for Graph Represent
se.set(rc={'figure.figsize':(7,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status' , 'Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
se.set(rc={'figure.figsize':(6,5)})
se.barplot(data=sales_state, x='Marital_Status', y='Amount' , hue='Gender')
```

```python
#] from the above gaph we can se that most of the buyers are Married (Womens) & they have High purchasing power
```

```python
# Product Category wise Sales

se.set(rc={'figure.figsize':(30,10)})
ax= se.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# product category wise orders

sales_state= df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(6)
se.set(rc={'figure.figsize':(15,10)})
se.barplot(data=sales_state, x='Product_Category', y='Amount')
```

```python
# From the above chart we got that the most sold products are 'Food', 'Clothing & Apparel', 'Electronics & Gadgets'
```
